```python
import re


class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next
class solution:
    def reverse(self, head: ListNode, left: int, right: int):
        if  not head or left == right:
            return head
        dummy = ListNode(val = 0, next = head)
        pre = dummy
        for _ in range(left-1):
           pre = pre.next
        cur = pre.next
        for _ in range(right-left):
            tmp = cur.next
            cur.next =  tmp.next
            tmp.next = pre.next
            pre.next = tmp

        return dummy.next
# Helper to build a linked list from a Python list
def build_list(arr):
    dummy = ListNode(0)
    cur = dummy
    for x in arr:
        cur.next = ListNode(x)
        cur = cur.next
    return dummy.next

# Helper to convert a linked list to Python list
def to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

head1 = build_list([1, 2, 3, 4, 5])
print(to_list(solution().reverse(head1, 2, 4)))



```

```python
from typing import List, Counter


class Solution:
    def frequencies(self, nums: List[int]):
        fre = Counter(nums)
        max_fre = max(fre.values())
        s = 0
        # print(fre, nums, max_fre)
        for i, e in enumerate(fre.values()):
            # print(i, e)
            if e == max_fre:
                s+=e
        return s
sol = Solution()
print(sol.frequencies([1,2,2,3,1,4]))
print(sol.frequencies(([1,2,3,4,5])))

```
