```python
def merge(nums: list[list[int]]):
  nums = sorted(nums, key = lambda x: x[0])
  stk = [nums[0]] 
  for i in range(1, len(nums)):
    el, er = stk[-1]
    il, ir = nums[i]
    # print(el, er)
    if il<=er:
      stk[-1] = [el, max(er, ir)]
    else:
      stk.append(nums[i])
  return stk 
print(merge([[1,3], [2,6], [8,10], [15,18]]))
print(merge([[1,4], [4,5]]))


```
